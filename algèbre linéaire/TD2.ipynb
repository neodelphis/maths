# Multiplication matricielle et produit scalaire

## Produit scalaire
$$
X =
\begin{bmatrix}
1\\ 
2\\ 
3 \\
4
\end{bmatrix}
$$

Transposée:
$$
X^T =
\begin{bmatrix}
1 & 2 & 3 & 4
\end{bmatrix}
$$

Calculez:
$$
X^T.X
\tag {1}
$$


# Multiplication matricielle

$$
\begin{bmatrix}
-2 & 1 & 0
\end{bmatrix}
\begin{bmatrix}
1 & 2 & 0\\ 
2 & 0 & 3\\ 
4 & 1 & 1
\end{bmatrix}
\tag {2}
$$

$$
\begin{bmatrix}
1 & 0 & 0\\
-2 & 1 & 0\\
-4 & 0 & 1\\

2 & 0 & 3\\ 
4 & 1 & 1
\end{bmatrix}
\tag {3}
$$

\end{bmatrix}
\tag {4}
$$
